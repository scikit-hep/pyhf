```python
%load_ext autoreload
%autoreload 2
import pyhf
import numpy as np
from pyhf.contrib.viz.brazil import plot_results
import matplotlib.pyplot as plt
```

```python
m = pyhf.simplemodels.hepdata_like([7.],[50.],[7.])
d = m.expected_data([0.0,1.0])
```

```python
mutests = np.linspace(0,5,11)
results = [pyhf.infer.hypotest(v,d,m,return_expected_set=True,
                               clip = True,
                               inclusive_pvalue = False,
                               calctype='asymptotics'
                              ) for v in mutests]  
toyresults = [pyhf.infer.hypotest(v,d,m,return_expected_set=True,
                                  inclusive_pvalue = False
                                  calctype='toybased',
                                 ) for v in mutests]  
```

```python
f,ax = plt.subplots()
plot_results(ax,mutests,toyresults, alpha = 0.1, colors = ['blue','red'])
plt.show()
f,ax = plt.subplots()
plot_results(ax,mutests,results, alpha = 0.2)
plt.show()
```
